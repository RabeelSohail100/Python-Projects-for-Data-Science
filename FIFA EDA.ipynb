```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Fifa = pd.read_csv('players_20.csv')
```

```python
Fifa.head()
```

```python
Fifa.shape
```

```python
for col in Fifa.columns:
    print(col)
```

```python
Fifa['nationality'].value_counts()
```

```python
Total_count = Fifa['nationality'].value_counts()[0:5]
```

```python
Keys = Fifa['nationality'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(6,6))
plt.bar(Keys,Total_count,color = ['Green','Blue','yellow','orange','red'])
plt.xlabel('States')
plt.ylabel('No of matches')
plt.title('Top 5 States Fifa Matches Analysis')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(Fifa['nationality'].value_counts().keys()[0:5]),list(Fifa['nationality'].value_counts()[0:5]),color = ['Green','Blue','yellow','orange','red'])
plt.xlabel('States')
plt.ylabel('No of matches')
plt.title('Top 5 States Fifa Matches Analysis')
plt.show()
```
